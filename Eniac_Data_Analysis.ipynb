```python
#
import matplotlib.pyplot as plt

```

```python
orders_qu['created_date'] = pd.to_datetime(orders_qu['created_date'])

# Set 'created_date' as the DataFrame's index
orders_qu.set_index('created_date', inplace=True)

# Calculate the sum of 'total_paid' on a monthly basis
monthly_sales = orders_qu['total_paid'].resample('M').sum()
```

```python
#Plot the resampled data to visualize the seasonal pattern:

import seaborn as sns
import matplotlib.pyplot as plt



# Create a bar plot with Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, color='g')

# Customize labels and title
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Monthly Sales with Seasonal Pattern')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.grid(axis='y')  # Add gridlines to the y-axis only
plt.tight_layout()
plt.show()

```

```python
import pandas as pd

# Assuming you have a DataFrame named 'data' with columns: 'product_id', 'order_date', 'quantity_sold', 'revenue', etc.

final_cl_qu_merged_with_categories.reset_index(inplace=True)

# Define Date Range 1
start_date_1 = '2017-11-13'
end_date_1 = '2017-11-19'

# Define Date Range 2
start_date_2 = '2017-01-13'
end_date_2 = '2017-01-19'

# Filter data for Date Range 1
data_range_1 = final_cl_qu_merged_with_categories[(final_cl_qu_merged_with_categories['date'] >= start_date_1) & (final_cl_qu_merged_with_categories['date'] <= end_date_1)]

# Filter data for Date Range 2
data_range_2 = final_cl_qu_merged_with_categories[(final_cl_qu_merged_with_categories['date'] >= start_date_2) & (final_cl_qu_merged_with_categories['date'] <= end_date_2)]

# Calculate product performance metrics for each date range
performance_range_1 = data_range_1.groupby('sku').agg({
    'product_quantity': 'sum',
    'total_paid': 'sum'
}).reset_index()

performance_range_2 = data_range_2.groupby('sku').agg({
     'product_quantity': 'sum',
    'total_paid': 'sum'
}).reset_index()

# Compare product performance between the two date ranges
performance_comparison = performance_range_1.merge(performance_range_2, on='sku', suffixes=('_range_1', '_range_2'))

# Calculate differences or percentages, for example:
performance_comparison['quantity_sold_change'] = (
    (performance_comparison['quantity_sold_range_1'] - performance_comparison['quantity_sold_range_2']) /
    performance_comparison['quantity_sold_range_2'] * 100
)

performance_comparison['revenue_change'] = (
    (performance_comparison['revenue_range_1'] - performance_comparison['revenue_range_2']) /
    performance_comparison['revenue_range_2'] * 100
)


# Display the performance comparison DataFrame
print(performance_comparison)


```

# Analyze discounts

1. Merge Products and Orderlines Tables

```python
# Select columns from "products_cl" and "orderlines_qu"
products_cl_selected = products_cl[['sku', 'name', 'price']]
orderlines_qu_selected = orderlines_qu[['id_order', 'unit_price', 'product_quantity', 'date', 'sku', 'id']]

# Merge the selected columns from both DataFrames on the 'sku' column
merged_ol_pdts = pd.merge(products_cl_selected, orderlines_qu_selected, on='sku', how='inner')

merged_ol_pdts["discount"]= merged_ol_pdts['price'] - merged_ol_pdts['unit_price']

merged_ol_pdts
```
